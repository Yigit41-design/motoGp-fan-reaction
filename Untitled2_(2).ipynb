## Build MotoGP Race Results Table (2019–2024)

In this step, we:

- Download the official MotoGP World Championship pages on Wikipedia for the 2019–2024 seasons.
- Extract and clean the race **calendar** (round, date, Grand Prix, circuit) and **results** (winning rider) tables for each year.
- Merge calendar and results on the **Round** column and add the corresponding season year.
- Concatenate all seasons into a single `motogp` DataFrame, where each row represents one race with its date, location and winning rider.

## 1. Data Collection: MotoGP Race Results (2019–2024)

In this step, we automatically collected MotoGP race data from publicly available Wikipedia pages for the 2019–2024 MotoGP World Championship seasons.  
For each season, we:

- Sent an HTTP request to the corresponding Wikipedia page (e.g., `2019_MotoGP_World_Championship`).
- Parsed all HTML tables on the page using `pandas.read_html`.
- Identified the **results table** containing the `Winning rider` information and cleaned the `Round` column.
- Identified the **calendar table** containing `Round`, `Date`, `Grand Prix`, and `Circuit` information.
- Merged the calendar and results tables on the `Round` column.
- Added a `Year` column to indicate the season and concatenated all seasons into a single `motogp` DataFrame.

The resulting `motogp` dataset contains one row per race, including the race year, round, date, Grand Prix name, circuit, and the winning rider.


```python
import requests
import pandas as pd
import warnings
warnings.filterwarnings("ignore", category=FutureWarning)
# Show all important columns clearly
pd.set_option("display.max_columns", 20)
pd.set_option("display.width", 200)

BASE = "https://en.wikipedia.org/wiki/"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0 Safari/537.36"
    )
}


def get_year_races(year):
    url = f"{BASE}{year}_MotoGP_World_Championship"

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()

    # Get all tables on the page
    tables = pd.read_html(resp.text)

    # 1) Find the table that contains "Winning rider"
    results_df = None
    for t in tables:
        cols = t.columns

        # Flatten MultiIndex if needed
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        if any("Winning rider" in c for c in flat_cols):
            results_df = t.copy()
            results_df.columns = flat_cols
            break

    if results_df is None:
        return None

    # Clean Round column (keep only numbers)
    if "Round" in results_df.columns:
        results_df["Round"] = results_df["Round"].astype(str).str.extract(r"(\d+)")
    else:
        return None

    # Keep only needed columns
    keep_cols_res = ["Round", "Winning rider"]
    keep_cols_res = [c for c in keep_cols_res if c in results_df.columns]
    results_df = results_df[keep_cols_res]

    # 2) Find the calendar table (Round, Date, Grand Prix, Circuit)
    calendar_df = None
    for t in tables:
        cols = t.columns
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        if ("Round" in flat_cols) and ("Date" in flat_cols) and ("Grand Prix" in flat_cols):
            tmp = t.copy()
            tmp.columns = flat_cols
            if "Circuit" in tmp.columns:
                calendar_df = tmp[["Round", "Date", "Grand Prix", "Circuit"]].copy()
            else:
                calendar_df = tmp[["Round", "Date", "Grand Prix"]].copy()
            break

    if calendar_df is None:
        return None

    # Clean Round column
    calendar_df["Round"] = calendar_df["Round"].astype(str).str.extract(r"(\d+)")

    # 3) Merge on Round
    merged = pd.merge(calendar_df, results_df, on="Round", how="inner")

    # Add year
    merged["Year"] = year

    # Order columns
    col_order = ["Year", "Round", "Date", "Grand Prix", "Circuit", "Winning rider"]
    col_order = [c for c in col_order if c in merged.columns]
    merged = merged[col_order]

    return merged


# Fetch races for 2019–2024
years = [2019, 2020, 2021, 2022, 2023, 2024]
all_races = []

for y in years:
    df_y = get_year_races(y)
    if df_y is not None:
        all_races.append(df_y)

if not all_races:
    raise ValueError("No tables found for any year. Please check the code/links.")

motogp = pd.concat(all_races, ignore_index=True)

# Show only the final table as output
motogp
```

## Extract Unique Winning Riders and Retrieve Their Nationalities

In this section, we:

- Handle exceptions using:

## 2. Unique Winning Riders & Nationality Extraction

In this step, we identified all **unique riders** who won at least one race between 2019–2024.  
After cleaning the rider names (removing footnotes like `[1]`), we created a list of distinct winners.

Then, for each rider, we scraped their **nationality** from their Wikipedia infobox.  
Special cases (e.g., page name differences) were handled manually.  
If nationality was missing, we inferred it from the `Born` field.

This produced the `rider_country_df` table containing each winning rider and their corresponding country.


```python
# 1) Önce kazanan pilotların benzersiz listesini görelim
unique_riders = (
    motogp["Winning rider"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)  # [1] gibi footnote'ları sil
    .str.strip()
    .unique()
)

print("Toplam benzersiz kazanan pilot sayısı:", len(unique_riders))
for r in sorted(unique_riders):
    print("-", r)
import re
from bs4 import BeautifulSoup
import requests

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0 Safari/537.36"
    )
}

special_pages = {
    "Jack Miller": "Jack_Miller_(motorcyclist)",
}

manual_nationality = {
    "Jack Miller": "Australia",   # ekstra garanti
}

def get_rider_nationality(name):
    # [1], [2] gibi footnote’ları sil
    clean = re.sub(r"\[.*?\]", "", str(name)).strip()

    # Eğer manuel nationality varsa direkt dön
    if clean in manual_nationality:
        return manual_nationality[clean]

    # Özel sayfa varsa onu, yoksa normal slug’ı kullan
    slug = special_pages.get(clean, clean.replace(" ", "_"))
    url = f"https://en.wikipedia.org/wiki/{slug}"

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()

    soup = BeautifulSoup(resp.text, "html.parser")
    infobox = soup.find("table", class_="infobox")
    if not infobox:
        return None

    for row in infobox.find_all("tr"):
        th = row.find("th")
        td = row.find("td")
        if not th or not td:
            continue

        label = th.get_text(strip=True)
        value = td.get_text(" ", strip=True)

        if label == "Nationality":
            return value
        # Bazı sayfalarda sadece "Born" satırında ülke geçebilir
        if label == "Born" and "," in value:
            parts = [p.strip() for p in value.split(",")]
            return parts[-1]  # son parça genelde ülke

    return None

# 1) rider_country_list'i ÖNCE oluştur
rider_country_list = []
# 2) loop BAŞLIYOR
for r in unique_riders:
    print("Çekiliyor:", r)
    c = get_rider_nationality(r)
    rider_country_list.append({"Name": r, "Country": c})

# 3) DataFrame'e çevir
rider_country_df = pd.DataFrame(rider_country_list)
rider_country_df
```

### Extracting Race Calendar and Results

This function retrieves the race schedule and results for each MotoGP season between 2019–2024 using Wikipedia pages.

For each year:
- It loads all tables on the page and identifies the **results table** containing the “Winning rider” column.
- It cleans the `Round` values and searches for an optional **Win margin** column.
- It then identifies the **calendar table**, which includes `Round`, `Date`, `Grand Prix`, and `Circuit`.
- Both tables are merged on `Round`, and the season year is added.

The output is a structured DataFrame containing race year, date, circuit, winning rider
ürüyor.


```python
def get_year_races(year):
    url = f"{BASE}{year}_MotoGP_World_Championship"
    print(f"\nYıl çekiliyor: {year}  -> {url}")

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()

    tables = pd.read_html(resp.text)
    print(f"  Toplam tablo sayısı: {len(tables)}")

    # 1) RESULTS / "Winning rider" tablosu
    results_df = None
    for i, t in enumerate(tables):
        cols = t.columns

        if isinstance(cols, pd.MultiIndex):
            flat_cols = [" ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                         for tup in cols]
        else:
            flat_cols = [str(c) for c in cols]

        if any("Winning rider" in c for c in flat_cols):
            print(f"  -> {year}: 'Winning rider' içeren tablo index {i} bulundu.")
            results_df = t.copy()
            results_df.columns = flat_cols
            break

    if results_df is None:
        print(f"  !!! {year} için 'Winning rider' içeren tablo bulunamadı!")
        return None

    # Round temizle
    if "Round" in results_df.columns:
        results_df["Round"] = results_df["Round"].astype(str).str.extract(r"(\d+)")
    else:
        print(f"  !!! {year} sonuç tablosunda 'Round' kolonu yok, atlıyorum.")
        return None

    # --- BURASI YENİ: Win margin kolonunu yakalamaya çalış ---
    win_margin_col = None
    for c in results_df.columns:
        lower = c.lower()
        if "winning margin" in lower or "win margin" in lower or "margin" in lower:
            win_margin_col = c
            break

    keep_cols_res = ["Round", "Winning rider"]
    if win_margin_col is not None:
        keep_cols_res.append(win_margin_col)

    keep_cols_res = [c for c in keep_cols_res if c in results_df.columns]
    results_df = results_df[keep_cols_res]

    # Kolon adını standartlaştır
    if win_margin_col is not None:
        results_df = results_df.rename(columns={win_margin_col: "Win margin"})

    # 2) Calendar tablosu
    calendar_df = None
    for i, t in enumerate(tables):
        cols = t.columns
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [" ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                         for tup in cols]
        else:
            flat_cols = [str(c) for c in cols]

        if ("Round" in flat_cols) and ("Date" in flat_cols) and ("Grand Prix" in flat_cols):
            tmp = t.copy()
            tmp.columns = flat_cols
            if "Circuit" in tmp.columns:
                calendar_df = tmp[["Round", "Date", "Grand Prix", "Circuit"]].copy()
            else:
                calendar_df = tmp[["Round", "Date", "Grand Prix"]].copy()
            print(f"  -> {year}: Calendar tablosu index {i} bulundu.")
            break

    if calendar_df is None:
        print(f"  !!! {year} için Calendar tablosu bulunamadı!")
        return None

    calendar_df["Round"] = calendar_df["Round"].astype(str).str.extract(r"(\d+)")

    merged = pd.merge(calendar_df, results_df, on="Round", how="inner")

    merged["Year"] = year

    col_order = ["Year", "Round", "Date", "Grand Prix", "Circuit", "Winning rider", "Win margin"]
    col_order = [c for c in col_order if c in merged.columns]
    merged = merged[col_order]

    print(f"  -> {year}: Toplam {len(merged)} yarış bulundu.")
    return merged
```

### Extracting Rider Career Information from Wikipedia

For each unique winning rider, we retrieved additional career information to enrich the dataset.  
This function cleans the rider’s name, resolves special cases (e.g., custom Wikipedia URLs), and then scrapes the rider’s Wikipedia page.

From the infobox, we extract:
- **Country**  
- **Years active / Active years**  
- **Championships**

We also scan the career statistics tables to obtain:
- **Career wins**
- **Career podiums**

When available, values are taken from the most complete row (e.g., “Total” or most recent MotoGP row).  
Manual overrides are applied for riders with inconsistent or non-standard page structures.

The result is a structured dictionary of rider-level information, later merged back into the main dataset.


```python
special_pages = {
    "Jack Miller": "Jack_Miller_(motorcyclist)",
    # gerekirse diğer özel sayfalar buraya
}

manual_overrides = {
    "Jack Miller": {
        "Country": "Australia"
    }
}

def get_rider_info(name):
    # footnote temizle
    clean = re.sub(r"\[.*?\]", "", str(name)).strip()

    info = {
        "Name": clean,
        "Country": None,
        "Years active": None,
        "Championships": None,
        "Career wins": None,
        "Career podiums": None,
    }

    # Manuel override varsa önce onları doldur
    if clean in manual_overrides:
        for k, v in manual_overrides[clean].items():
            if k in info:
                info[k] = v

    slug = special_pages.get(clean, clean.replace(" ", "_"))
    url = f"https://en.wikipedia.org/wiki/{slug}"

    print("  Wiki çekiliyor:", clean, "->", url)

    resp = requests.get(url, headers=headers)
    if not resp.ok:
        print("   !!! Sayfa alınamadı:", resp.status_code)
        return info

    soup = BeautifulSoup(resp.text, "html.parser")
    infobox = soup.find("table", class_="infobox")
    if infobox:
        for row in infobox.find_all("tr"):
            th = row.find("th")
            td = row.find("td")
            if not th or not td:
                continue

            label = th.get_text(strip=True)
            value = td.get_text(" ", strip=True)

            # Country
            if label == "Nationality" and info["Country"] is None:
                info["Country"] = value

            # Years active
            # Bazı sayfalarda "Years active", bazılarında "Active years" olabilir.
            if ("Years active" in label or "Active years" in label) and info["Years active"] is None:
                info["Years active"] = value

            # Championships
            if "Championships" in label and info["Championships"] is None:
                info["Championships"] = value

    # Career wins / podiums için istatistik tablolarına bakalım
    # Genelde "Grand Prix motorcycle racing career statistics" başlıklı bir tablo oluyor
    stats_tables = soup.find_all("table", class_="wikitable")
    for table in stats_tables:
        caption = table.find("caption")
        cap_text = caption.get_text(strip=True) if caption else ""
        if "career statistics" in cap_text.lower() or "grand prix motorcycle racing" in cap_text.lower():
            # başlık satırını al
            headers_row = table.find("tr")
            headers_text = [th.get_text(" ", strip=True).lower() for th in headers_row.find_all("th")]

            # Hangi kolonda wins/podiums var bul
            win_idx = None
            pod_idx = None
            for idx, ht in enumerate(headers_text):
                if "wins" in ht or "race wins" in ht:
                    win_idx = idx
                if "podiums" in ht:
                    pod_idx = idx

            # Son satırı (en güncel satır) alıp oradan veri almaya çalış
            rows = table.find_all("tr")[1:]  # header hariç
            if rows:
                last_row_tds = rows[-1].find_all(["td", "th"])
                if win_idx is not None and win_idx < len(last_row_tds) and info["Career wins"] is None:
                    info["Career wins"] = last_row_tds[win_idx].get_text(strip=True)
                if pod_idx is not None and pod_idx < len(last_row_tds) and info["Career podiums"] is None:
                    info["Career podiums"] = last_row_tds[pod_idx].get_text(strip=True)
            break  # bir tane bulduk, daha fazla gezmeyelim

    return info
```

### Cleaning Rider Names and Building the Rider Information Table

We first cleaned the `Winning rider` field by removing footnote markers (e.g., `[1]`) and trimming whitespace, creating a standardized `Winning rider_clean` column.

From this cleaned column, we extracted the list of **unique winning riders** across all seasons.  
For each rider, we applied the `get_rider_info()` function to collect nationality, career statistics, championships, and active years from their Wikipedia page.

All extracted information was stored in `rider_info_df`, a structured rider-level dataset that will later be merged with the race-level MotoGP data.


```python
# Winning rider'ı temizleyip ek bir kolon oluşturalım
motogp["Winning rider_clean"] = (
    motogp["Winning rider"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)

unique_riders = sorted(motogp["Winning rider_clean"].unique())
print("Toplam benzersiz kazanan pilot sayısı:", len(unique_riders))

rider_info_list = []
for r in unique_riders:
    print("Sürücü işleniyor:", r)
    info = get_rider_info(r)
    rider_info_list.append(info)

rider_info_df = pd.DataFrame(rider_info_list)
rider_info_df.head()
```

### Complete MotoGP Data Pipeline: Race Extraction, Rider Scraping, Feature Cleaning, and Popularity Scoring

This block builds the full MotoGP dataset used for analysis.  
It performs the following operations:

1. **Scrape race data (2019–2024):**  
   - Extract race calendars and winning riders from each season’s Wikipedia page.  
   - Capture race details such as round, date, Grand Prix, circuit, and win margin.  
   - Combine all seasons into a single race-level DataFrame.

2. **Collect rider information:**  
   - Clean winner names and identify all unique riders.  
   - For each rider, scrape their Wikipedia page to obtain:
     - Country  
     - Years active  
     - Championships  
     - Career wins  
     - Career podiums  

3. **Merge race-level and rider-level data:**  
   - Join rider information back into the race table to enrich each race with rider attributes.

4. **Convert rider attributes into numeric features:**  
   - Extract numerical values from text fields.  
   - Compute total years active from ranges such as “2008–2022” or “2013–present”.

5. **Compute a weighted popularity score:**  
   - Popularity =  
     `100 × Championships + 10 × Career Wins + 3 × Career Podiums + 1 × Years Active`  
   - Attach the popularity score to both rider-level and race-level tables.

6. **Generate a ranked popularity table:**  
   - Create `rider_popularity`, listing all riders sorted by their computed popularity score.

This block produces the final cleaned, enriched, and feature-engineered dataset used for all subsequent EDA and hypothesis testing.


```python
import re
import requests
import pandas as pd
from bs4 import BeautifulSoup

# Ekranda düzgün gözüksün
pd.set_option("display.max_columns", 30)
pd.set_option("display.width", 220)

BASE = "https://en.wikipedia.org/wiki/"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0 Safari/537.36"
    )
}

# ---------------------------------------------------------
# 1)
# ---------------------------------------------------------

def get_year_races(year):
    url = f"{BASE}{year}_MotoGP_World_Championship"
    print(f"\nYıl çekiliyor: {year}  -> {url}")

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()

    tables = pd.read_html(resp.text)
    print(f"  Toplam tablo sayısı: {len(tables)}")

    # --- 1) RESULTS / "Winning rider" tablosu ---
    results_df = None
    for i, t in enumerate(tables):
        cols = t.columns

        # MultiIndex başlıkları düzleştir
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        # "Winning rider" içeren tabloyu bul
        if any("Winning rider" in c for c in flat_cols):
            print(f"  -> {year}: 'Winning rider' içeren tablo index {i} bulundu.")
            results_df = t.copy()
            results_df.columns = flat_cols
            break

    if results_df is None:
        print(f"  !!! {year} için 'Winning rider' içeren tablo bulunamadı!")
        return None

    # Round temizle
    if "Round" in results_df.columns:
        results_df["Round"] = results_df["Round"].astype(str).str.extract(r"(\d+)")
    else:
        print(f"  !!! {year} sonuç tablosunda 'Round' kolonu yok, atlıyorum.")
        return None

    # Win margin kolonunu yakalamaya çalış
    win_margin_col = None
    for c in results_df.columns:
        lower = c.lower()
        if (
            "winning margin" in lower
            or "win margin" in lower
            or ("margin" in lower and "time" not in lower)
        ):
            win_margin_col = c
            break

    keep_cols_res = ["Round", "Winning rider"]
    if win_margin_col is not None:
        keep_cols_res.append(win_margin_col)

    keep_cols_res = [c for c in keep_cols_res if c in results_df.columns]
    results_df = results_df[keep_cols_res]

    if win_margin_col is not None:
        results_df = results_df.rename(columns={win_margin_col: "Win margin"})

    # --- 2) Takvim tablosu (Round, Date, Grand Prix, Circuit) ---
    calendar_df = None
    for i, t in enumerate(tables):
        cols = t.columns
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        if ("Round" in flat_cols) and ("Date" in flat_cols) and ("Grand Prix" in flat_cols):
            tmp = t.copy()
            tmp.columns = flat_cols
            if "Circuit" in tmp.columns:
                calendar_df = tmp[["Round", "Date", "Grand Prix", "Circuit"]].copy()
            else:
                calendar_df = tmp[["Round", "Date", "Grand Prix"]].copy()
            print(f"  -> {year}: Calendar tablosu index {i} bulundu.")
            break

    if calendar_df is None:
        print(f"  !!! {year} için Calendar tablosu bulunamadı!")
        return None

    calendar_df["Round"] = calendar_df["Round"].astype(str).str.extract(r"(\d+)")

    # --- 3) Birleştir ---
    merged = pd.merge(calendar_df, results_df, on="Round", how="inner")
    merged["Year"] = year

    col_order = [
        "Year",
        "Round",
        "Date",
        "Grand Prix",
        "Circuit",
        "Winning rider",
        "Win margin",
    ]
    col_order = [c for c in col_order if c in merged.columns]
    merged = merged[col_order]

    print(f"  -> {year}: Toplam {len(merged)} yarış bulundu.")
    return merged


# ---------------------------------------------------------
# 2) SÜRÜCÜ BİLGİLERİ SCRAPER'I
# ---------------------------------------------------------

special_pages = {
    "Jack Miller": "Jack_Miller_(motorcyclist)",
    # başka istisnalar olursa buraya ekleyebilirsin
}

manual_overrides = {
    "Jack Miller": {"Country": "Australia"},
    # istersen bazı sürücüler için wins/podiums'u da buraya elle yazabilirsin
}


def _extract_career_stats_from_tables(soup, info):
    """
    Sayfadaki TÜM tablolara bakar;
    başlık satırında hem 'win' hem 'podium' geçen tabloyu bulmaya çalışır.
    Öncelik:
      - 'Total' satırı
      - yoksa 'MotoGP' satırı
      - o da yoksa son satır
    Buradan Career wins / podiums alanlarını doldurur.
    """
    # tüm tabloları tara (infobox içindekiler dâhil)
    for table in soup.find_all("table"):
        header_row = table.find("tr")
        if not header_row:
            continue

        header_cells = header_row.find_all(["th", "td"])
        if len(header_cells) < 2:
            continue

        headers_text = [
            cell.get_text(" ", strip=True).lower() for cell in header_cells
        ]

        # Hem win hem podium sütunu olmalı
        if not (any("win" in h for h in headers_text) and any("podium" in h for h in headers_text)):
            continue

        win_idx = None
        pod_idx = None
        for idx, ht in enumerate(headers_text):
            if win_idx is None and ("wins" in ht or "win" in ht):
                win_idx = idx
            if pod_idx is None and "podium" in ht:
                pod_idx = idx

        if win_idx is None and pod_idx is None:
            continue

        rows = table.find_all("tr")[1:]  # başlık hariç
        if not rows:
            continue

        chosen_row = None

        # Önce "Total" satırına bak
        for row in rows:
            first = row.find(["th", "td"])
            if first and "total" in first.get_text(strip=True).lower():
                chosen_row = row
                break

        # Sonra özellikle "MotoGP" satırını dene
        if chosen_row is None:
            for row in rows:
                first = row.find(["th", "td"])
                txt = first.get_text(strip=True).lower() if first else ""
                if "motogp" in txt:
                    chosen_row = row
                    break

        # Hâlâ yoksa son satırı al
        if chosen_row is None:
            chosen_row = rows[-1]

        cells = chosen_row.find_all(["td", "th"])

        if win_idx is not None and win_idx < len(cells) and info["Career wins"] is None:
            val = cells[win_idx].get_text(" ", strip=True)
            info["Career wins"] = val or None

        if pod_idx is not None and pod_idx < len(cells) and info["Career podiums"] is None:
            val = cells[pod_idx].get_text(" ", strip=True)
            info["Career podiums"] = val or None

        # En az birini doldurduysak yeter
        if info["Career wins"] is not None or info["Career podiums"] is not None:
            break


def get_rider_info(name):
    # footnote temizle
    clean = re.sub(r"\[.*?\]", "", str(name)).strip()

    info = {
        "Name": clean,
        "Country": None,
        "Years active": None,
        "Championships": None,
        "Career wins": None,
        "Career podiums": None,
    }

    # Manuel override'lar
    if clean in manual_overrides:
        for k, v in manual_overrides[clean].items():
            if k in info:
                info[k] = v

    slug = special_pages.get(clean, clean.replace(" ", "_"))
    url = f"https://en.wikipedia.org/wiki/{slug}"

    print("  Wiki çekiliyor:", clean, "->", url)

    resp = requests.get(url, headers=headers)
    if not resp.ok:
        print("   !!! Sayfa alınamadı:", resp.status_code)
        return info

    soup = BeautifulSoup(resp.text, "html.parser")

    # ---- İnfobox: Country / Years active / Championships ----
    infobox = soup.find("table", class_="infobox")
    if infobox:
        for row in infobox.find_all("tr"):
            th = row.find("th")
            td = row.find("td")
            if not th or not td:
                continue

            label = th.get_text(strip=True)
            value = td.get_text(" ", strip=True)

            # Country
            if label == "Nationality" and info["Country"] is None:
                info["Country"] = value

            # Bazı sayfalarda Nationality yok, ülke doğum bilgisinin son parçasında
            if label == "Born" and info["Country"] is None and "," in value:
                parts = [p.strip() for p in value.split(",")]
                info["Country"] = parts[-1]

            # Years active / Active years
            if ("Years active" in label or "Active years" in label) and info["Years active"] is None:
                info["Years active"] = value

            # Championships
            if "Championships" in label and info["Championships"] is None:
                info["Championships"] = value

    # ---- Kariyer istatistik tablosu: career wins / podiums ----
    _extract_career_stats_from_tables(soup, info)

    return info


# ---------------------------------------------------------
# 3) HER ŞEYİ ÇALIŞTIR, BİRLEŞTİR
# ---------------------------------------------------------

# Yıllar
years = [2019, 2020, 2021, 2022, 2023, 2024]

all_races = []
for y in years:
    df_y = get_year_races(y)
    if df_y is not None:
        all_races.append(df_y)

if not all_races:
    raise ValueError("Hiçbir yıl için tablo bulunamadı, kodu/bağlantıları kontrol et 🙃")

motogp = pd.concat(all_races, ignore_index=True)
print("\nTOPLAM YARIŞ SAYISI:", len(motogp))

# Kazanan isimleri temizleyelim
motogp["Winning rider_clean"] = (
    motogp["Winning rider"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)

unique_riders = sorted(motogp["Winning rider_clean"].unique())
print("Toplam benzersiz kazanan pilot sayısı:", len(unique_riders))

rider_info_list = []
for r in unique_riders:
    print("Sürücü işleniyor:", r)
    info = get_rider_info(r)
    rider_info_list.append(info)

rider_info_df = pd.DataFrame(rider_info_list)

# Son tabloyu merge edelim
motogp_full = motogp.merge(
    rider_info_df,
    left_on="Winning rider_clean",
    right_on="Name",
    how="left",
)

motogp_full = motogp_full.drop(columns=["Winning rider_clean", "Name"])

# Sütun sırasını ayarlayalım
col_order = [
    "Year",
    "Date",
    "Grand Prix",
    "Circuit",
    "Winning rider",
    "Win margin",
    "Career wins",
    "Career podiums",
    "Championships",
    "Country",
    "Years active",
]

col_order = [c for c in col_order if c in motogp_full.columns]
motogp_full = motogp_full[col_order]

motogp_full.head()
rider_info_df = pd.DataFrame(rider_info_list)

# ---------------------------------------------------------
# 4) SAYISAL KOLONLARI TEMİZLE + POPULARITY HESAPLA
# ---------------------------------------------------------

def _to_int_first_number(x):
    """'7 (2013, 2014, ...)' gibi stringlerden ilk gördüğü sayıyı çeker.
       None veya sayı yoksa 0 döner."""
    if pd.isna(x):
        return 0
    s = str(x)
    m = re.search(r"(\d+)", s)
    return int(m.group(1)) if m else 0


def _years_active_length(s, current_year=2024):
    """'2008–2022', '2013–present', '2001–2010, 2012–present' gibi
       stringlerden toplam kapsanan yıl sayısını yaklaşık hesaplar."""
    if pd.isna(s):
        return 0
    s = str(s)

    years = [int(y) for y in re.findall(r"(\d{4})", s)]
    if not years:
        return 0

    # present varsa onu current_year olarak düşün
    # örn: '2013–present' -> [2013] gelir, ama stringde 'present' var
    if "present" in s.lower():
        years.append(current_year)

    return max(years) - min(years) + 1


# Sayısal versiyonları ekleyelim
rider_info_df["Career wins_num"] = rider_info_df["Career wins"].apply(_to_int_first_number)
rider_info_df["Career podiums_num"] = rider_info_df["Career podiums"].apply(_to_int_first_number)
rider_info_df["Championships_num"] = rider_info_df["Championships"].apply(_to_int_first_number)
rider_info_df["Years_active_len"] = rider_info_df["Years active"].apply(_years_active_length)

# Ağırlıklar (istediğin zaman değiştirebilirsin)
W_CHAMP = 100
W_WIN = 10
W_POD = 3
W_YEARS = 1

rider_info_df["Popularity"] = (
    W_CHAMP * rider_info_df["Championships_num"]
    + W_WIN * rider_info_df["Career wins_num"]
    + W_POD * rider_info_df["Career podiums_num"]
    + W_YEARS * rider_info_df["Years_active_len"]
)

# ---------------------------------------------------------
# 5) YARIŞ TABLOSUNA POPULARITY EKLE
# ---------------------------------------------------------

motogp_full = motogp.merge(
    rider_info_df,
    left_on="Winning rider_clean",
    right_on="Name",
    how="left",
)

motogp_full = motogp_full.drop(columns=["Winning rider_clean", "Name"])

col_order = [
    "Year",
    "Date",
    "Grand Prix",
    "Circuit",
    "Winning rider",
    "Win margin",
    "Career wins",
    "Career podiums",
    "Championships",
    "Country",
    "Years active",
    "Popularity",           # yeni kolon
]

col_order = [c for c in col_order if c in motogp_full.columns]
motogp_full = motogp_full[col_order]

motogp_full.head()
# Her sürücü için popularity tablosu
rider_popularity = (
    rider_info_df[
        ["Name",
         "Popularity",
         "Championships_num",
         "Career wins_num",
         "Career podiums_num",
         "Years_active_len"]
    ]
    .sort_values("Popularity", ascending=False)
    .reset_index(drop=True)
)

rider_popularity
```

### MotoGP race–level dataset with rider stats and Google Trends

In this code block we build the **final analysis dataset**: a race-level table that combines MotoGP race results (2019–2024), rider career statistics, a custom popularity score, and Google search interest before and after each race.

- **Imports & global setup:**  
  Load web-scraping, data-handling and time libraries (`requests`, `pandas`, `BeautifulSoup`, `pytrends`, `datetime`), set pandas display options, define the base Wikipedia URL, HTTP headers, and the `TrendReq` client.

- **Scrape races from Wikipedia (`get_year_races`)**  
  For each season page (2019–2024), read all HTML tables, find the results table that contains **“Winning rider”**, clean the `Round` column, optionally capture **win margin**, and find the calendar table with **Round–Date–Grand Prix–Circuit**. Merge these tables by `Round` and add a `Year` column.

- **Parse real race date (`parse_race_date`)**  
  Convert messy Wikipedia date strings (e.g. `"7–9 June"`, `"10 March"`) plus the `Year` into a single `RaceDate` `datetime` object by taking the **last day** in the range.

- **Scrape rider info from Wikipedia (`get_rider_info`)**  
  Define `special_pages` and `manual_overrides` for riders with non-standard wiki URLs / fixed data (e.g. Jack Miller). For each rider, open their page, read the infobox and extract **Country**, **Years active / Active years**, and **Championships**. Then scan tables whose headers contain both “win” and “podium” to pull **career wins** and **career podiums**.

- **Build base race & rider tables**  
  Loop over all years, call `get_year_races`, and concatenate into a single `motogp` DataFrame. Clean `Winning rider` into `Winning rider_clean`, create the unique rider list, loop over riders with `get_rider_info`, and collect everything into `rider_info_df`.

- **Engineer numeric career features & popularity score**  
  Define helpers to:  
  - extract the **first integer** from strings like `"7 (2013, 2014, ...)"`,  
  - approximate total **active years** from ranges like `"2008–2022"` or `"2013–present"`.  
  Create numeric columns (`Career wins_num`, `Career podiums_num`, `Championships_num`, `Years_active_len`) and compute a **Popularity** index as a weighted sum of titles, wins, podiums and years.

- **Merge rider features back to races**  
  Join `motogp` and `rider_info_df` on `Winning rider_clean = Name` to form `motogp_full`, drop helper columns, and add the `RaceDate` column using `parse_race_date` row-wise.

- **Google Trends before/after interest (`get_search_interest`)**  
  Define a function that, for each race and winning rider, queries Google Trends in two windows:  
  – **7 days before** the race,  
  – **7 days after** the race.  
  It returns average search levels in both windows, their **difference**, and the **relative increase percentage**.

- **Populate search-interest features & finalize table**  
  Loop over all races in `motogp_full`, call `get_search_interest`, and fill `Search Before`, `Search After`, `Trend Difference`, and `Relative Increase %`. Finally, reorder the columns into a clean schema and print the full `motogp_full` table, which is the main dataset used for EDA and hypothesis testing.

```python

import re
import time
from datetime import datetime, timedelta
import pandas as pd
import requests
from bs4 import BeautifulSoup
from pytrends.request import TrendReq
# ---------------------------------------------------------
# GENEL AYARLAR
# ---------------------------------------------------------

pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 240)

BASE = "https://en.wikipedia.org/wiki/"

headers = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 (KHTML, like Gecko) "
        "Chrome/120.0 Safari/537.36"
    )
}

# Google Trends client (biraz geniş timeout)
pytrends = TrendReq(
    hl="en-US",
    tz=360,
    timeout=(10, 30)
)

# ---------------------------------------------------------
# 1) HER YIL İÇİN YARIŞ TAKVİMİ + KAZANAN + WIN MARGIN
# ---------------------------------------------------------

def get_year_races(year: int) -> pd.DataFrame | None:
    url = f"{BASE}{year}_MotoGP_World_Championship"
    print(f"\nYıl çekiliyor: {year}  -> {url}")

    resp = requests.get(url, headers=headers)
    resp.raise_for_status()

    tables = pd.read_html(resp.text)
    print(f"  Toplam tablo sayısı: {len(tables)}")

    # ---------- 1) RESULTS / "Winning rider" tablosu ----------
    results_df = None
    for i, t in enumerate(tables):
        cols = t.columns

        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        if any("Winning rider" in c for c in flat_cols):
            print(f"  -> {year}: 'Winning rider' içeren tablo index {i} bulundu.")
            results_df = t.copy()
            results_df.columns = flat_cols
            break

    if results_df is None:
        print(f"  !!! {year} için 'Winning rider' içeren tablo bulunamadı!")
        return None

    if "Round" in results_df.columns:
        results_df["Round"] = results_df["Round"].astype(str).str.extract(r"(\d+)")
    else:
        print(f"  !!! {year} sonuç tablosunda 'Round' kolonu yok, atlıyorum.")
        return None

    # Win margin kolonunu yakala
    win_margin_col = None
    for c in results_df.columns:
        lower = c.lower()
        if (
            "winning margin" in lower
            or "win margin" in lower
            or ("margin" in lower and "time" not in lower)
        ):
            win_margin_col = c
            break

    keep_cols_res = ["Round", "Winning rider"]
    if win_margin_col is not None:
        keep_cols_res.append(win_margin_col)

    keep_cols_res = [c for c in keep_cols_res if c in results_df.columns]
    results_df = results_df[keep_cols_res]

    if win_margin_col is not None:
        results_df = results_df.rename(columns={win_margin_col: "Win margin"})

    # ---------- 2) Takvim tablosu ----------
    calendar_df = None
    for i, t in enumerate(tables):
        cols = t.columns
        if isinstance(cols, pd.MultiIndex):
            flat_cols = [
                " ".join([str(x) for x in tup if str(x) != "nan"]).strip()
                for tup in cols
            ]
        else:
            flat_cols = [str(c) for c in cols]

        if ("Round" in flat_cols) and ("Date" in flat_cols) and ("Grand Prix" in flat_cols):
            tmp = t.copy()
            tmp.columns = flat_cols
            if "Circuit" in tmp.columns:
                calendar_df = tmp[["Round", "Date", "Grand Prix", "Circuit"]].copy()
            else:
                calendar_df = tmp[["Round", "Date", "Grand Prix"]].copy()
            print(f"  -> {year}: Calendar tablosu index {i} bulundu.")
            break

    if calendar_df is None:
        print(f"  !!! {year} için Calendar tablosu bulunamadı!")
        return None

    calendar_df["Round"] = calendar_df["Round"].astype(str).str.extract(r"(\d+)")

    # ---------- 3) Birleştir ----------
    merged = pd.merge(calendar_df, results_df, on="Round", how="inner")
    merged["Year"] = year

    col_order = [
        "Year",
        "Round",
        "Date",
        "Grand Prix",
        "Circuit",
        "Winning rider",
        "Win margin",
    ]
    col_order = [c for c in col_order if c in merged.columns]
    merged = merged[col_order]

    print(f"  -> {year}: Toplam {len(merged)} yarış bulundu.")
    return merged

# ---------------------------------------------------------
# 1.1) WIKI'DEKİ DATE + YEAR'DAN GERÇEK RACE DATE ÜRET
# ---------------------------------------------------------

MONTH_MAP = {
    "january": 1,
    "february": 2,
    "march": 3,
    "april": 4,
    "may": 5,
    "june": 6,
    "july": 7,
    "august": 8,
    "september": 9,
    "october": 10,
    "november": 11,
    "december": 12,
}

def parse_race_date(date_str: str, year: int) -> datetime | None:
    if pd.isna(date_str):
        return None

    s = str(date_str)
    s = re.sub(r"\[.*?\]", "", s)  # footnote sil
    s = s.strip()

    tokens = s.split()
    if len(tokens) == 1:
        month_name = tokens[0].lower()
        month = MONTH_MAP.get(month_name, 1)
        day = 1
    else:
        day_part = tokens[0]
        month_name = tokens[-1].lower()
        month = MONTH_MAP.get(month_name, 1)

        m = re.findall(r"\d+", day_part)
        if not m:
            day = 1
        else:
            day = int(m[-1])  # son sayı

    try:
        return datetime(year, month, day)
    except ValueError:
        return None

# ---------------------------------------------------------
# 2) SÜRÜCÜ BİLGİLERİ SCRAPER'I
# ---------------------------------------------------------

special_pages = {
    "Jack Miller": "Jack_Miller_(motorcyclist)",
}

manual_overrides = {
    "Jack Miller": {"Country": "Australia"},
}

def _extract_career_stats_from_tables(soup, info: dict) -> None:
    for table in soup.find_all("table"):
        header_row = table.find("tr")
        if not header_row:
            continue

        header_cells = header_row.find_all(["th", "td"])
        if len(header_cells) < 2:
            continue

        headers_text = [cell.get_text(" ", strip=True).lower() for cell in header_cells]

        if not (any("win" in h for h in headers_text) and any("podium" in h for h in headers_text)):
            continue

        win_idx = None
        pod_idx = None
        for idx, ht in enumerate(headers_text):
            if win_idx is None and ("wins" in ht or "win" in ht):
                win_idx = idx
            if pod_idx is None and "podium" in ht:
                pod_idx = idx

        if win_idx is None and pod_idx is None:
            continue

        rows = table.find_all("tr")[1:]
        if not rows:
            continue

        chosen_row = None

        for row in rows:
            first = row.find(["th", "td"])
            if first and "total" in first.get_text(strip=True).lower():
                chosen_row = row
                break

        if chosen_row is None:
            for row in rows:
                first = row.find(["th", "td"])
                txt = first.get_text(strip=True).lower() if first else ""
                if "motogp" in txt:
                    chosen_row = row
                    break

        if chosen_row is None:
            chosen_row = rows[-1]

        cells = chosen_row.find_all(["td", "th"])

        if win_idx is not None and win_idx < len(cells) and info["Career wins"] is None:
            val = cells[win_idx].get_text(" ", strip=True)
            info["Career wins"] = val or None

        if pod_idx is not None and pod_idx < len(cells) and info["Career podiums"] is None:
            val = cells[pod_idx].get_text(" ", strip=True)
            info["Career podiums"] = val or None

        if info["Career wins"] is not None or info["Career podiums"] is not None:
            break

def get_rider_info(name: str) -> dict:
    clean = re.sub(r"\[.*?\]", "", str(name)).strip()

    info = {
        "Name": clean,
        "Country": None,
        "Years active": None,
        "Championships": None,
        "Career wins": None,
        "Career podiums": None,
    }

    if clean in manual_overrides:
        for k, v in manual_overrides[clean].items():
            if k in info:
                info[k] = v

    slug = special_pages.get(clean, clean.replace(" ", "_"))
    url = f"https://en.wikipedia.org/wiki/{slug}"

    print("  Wiki çekiliyor:", clean, "->", url)

    resp = requests.get(url, headers=headers)
    if not resp.ok:
        print("   !!! Sayfa alınamadı:", resp.status_code)
        return info

    soup = BeautifulSoup(resp.text, "html.parser")

    infobox = soup.find("table", class_="infobox")
    if infobox:
        for row in infobox.find_all("tr"):
            th = row.find("th")
            td = row.find("td")
            if not th or not td:
                continue

            label = th.get_text(strip=True)
            value = td.get_text(" ", strip=True)

            if label == "Nationality" and info["Country"] is None:
                info["Country"] = value

            if label == "Born" and info["Country"] is None and "," in value:
                parts = [p.strip() for p in value.split(",")]
                info["Country"] = parts[-1]

            if ("Years active" in label or "Active years" in label) and info["Years active"] is None:
                info["Years active"] = value

            if "Championships" in label and info["Championships"] is None:
                info["Championships"] = value

    _extract_career_stats_from_tables(soup, info)

    return info

# ---------------------------------------------------------
# 3) TÜM YILLAR + RIDER INFO + POPULARITY
# ---------------------------------------------------------

# Burada hangi yıllara bakacağını belirlersin.
# İstersen sadece [2025] yapıp tek yıl çalıştırırsın.
years = [2019, 2020, 2021, 2022, 2023, 2024]

all_races = []
for y in years:
    df_y = get_year_races(y)
    if df_y is not None:
        all_races.append(df_y)

if not all_races:
    raise ValueError("Hiçbir yıl için tablo bulunamadı, kodu/bağlantıları kontrol et 🙃")

motogp = pd.concat(all_races, ignore_index=True)
print("\nTOPLAM YARIŞ SAYISI:", len(motogp))

motogp["Winning rider_clean"] = (
    motogp["Winning rider"]
    .astype(str)
    .str.replace(r"\[.*?\]", "", regex=True)
    .str.strip()
)

unique_riders = sorted(motogp["Winning rider_clean"].unique())
print("Toplam benzersiz kazanan pilot sayısı:", len(unique_riders))

rider_info_list = []
for r in unique_riders:
    print("Sürücü işleniyor:", r)
    info = get_rider_info(r)
    rider_info_list.append(info)

rider_info_df = pd.DataFrame(rider_info_list)

# ---------------------------------------------------------
# 3.1) SAYISAL KOLONLAR + POPULARITY
# ---------------------------------------------------------

def _to_int_first_number(x):
    if pd.isna(x):
        return 0
    s = str(x)
    m = re.search(r"(\d+)", s)
    return int(m.group(1)) if m else 0

def _years_active_length(s, current_year=None):
    if pd.isna(s):
        return 0
    s = str(s)

    if current_year is None:
        current_year = datetime.now().year

    years_found = [int(y) for y in re.findall(r"(\d{4})", s)]
    if not years_found:
        return 0

    if "present" in s.lower():
        years_found.append(current_year)

    return max(years_found) - min(years_found) + 1

rider_info_df["Career wins_num"] = rider_info_df["Career wins"].apply(_to_int_first_number)
rider_info_df["Career podiums_num"] = rider_info_df["Career podiums"].apply(_to_int_first_number)
rider_info_df["Championships_num"] = rider_info_df["Championships"].apply(_to_int_first_number)
rider_info_df["Years_active_len"] = rider_info_df["Years active"].apply(_years_active_length)

W_CHAMP = 100
W_WIN   = 10
W_POD   = 3
W_YEARS = 1

rider_info_df["Popularity"] = (
    W_CHAMP * rider_info_df["Championships_num"]
    + W_WIN * rider_info_df["Career wins_num"]
    + W_POD * rider_info_df["Career podiums_num"]
    + W_YEARS * rider_info_df["Years_active_len"]
)

# ---------------------------------------------------------
# 3.2) YARIŞ TABLOSUNA MERGE + RaceDate
# ---------------------------------------------------------

motogp_full = motogp.merge(
    rider_info_df,
    left_on="Winning rider_clean",
    right_on="Name",
    how="left",
)

motogp_full.drop(columns=["Winning rider_clean", "Name"], inplace=True)

motogp_full["RaceDate"] = motogp_full.apply(
    lambda row: parse_race_date(row["Date"], int(row["Year"])),
    axis=1
)

# ---------------------------------------------------------
# 4) GOOGLE TRENDS: PİLOTLAR İÇİN TEK SEFERDE SERİ ÇEK (YIL YIL)
# ---------------------------------------------------------

def safe_trend_series(name: str, start_date: datetime, end_date: datetime, max_retries: int = 5):
    timeframe = f"{start_date:%Y-%m-%d} {end_date:%Y-%m-%d}"
    for attempt in range(1, max_retries + 1):
        try:
            pytrends.build_payload([name], timeframe=timeframe)
            df = pytrends.interest_over_time()
            if df.empty:
                print(f"  Trend boş döndü: {name} ({timeframe})")
                return None
            df = df[[name]]  # sadece ilgili kolon kalsın
            return df
        except Exception as e:
            print(f"  Trend hata ({name}), deneme {attempt}/{max_retries}: {e}")
            # 429 riskini azaltmak için bekle
            sleep_sec = 5 * attempt
            print(f"   {sleep_sec} saniye bekleniyor...")
            time.sleep(sleep_sec)

    print(f"  >>> Trend verisi alınamadı (tamamen başarısız): {name}")
    return None

# Yılları hangi sırayla trends'e soralım?
# 2025'ten başla, sonra 2024, 2023 ... gibi.
YEAR_ORDER_FOR_TRENDS = [2025, 2024, 2023, 2022, 2021, 2020, 2019]

# rider_trend_data: (rider, year) -> dataframe
rider_trend_data = {}

print("\nGoogle Trends serileri çekiliyor (yıl yıl, pilot başına 1 istek)...")

for y in YEAR_ORDER_FOR_TRENDS:
    if y not in motogp_full["Year"].values:
        continue  # o yıl yoksa geç

    df_y = motogp_full[motogp_full["Year"] == y].copy()

    # O yılın tarih aralığı
    min_race_date = df_y["RaceDate"].min()
    max_race_date = df_y["RaceDate"].max()

    if pd.isna(min_race_date) or pd.isna(max_race_date):
        print(f"\n{y} için geçerli RaceDate yok, trends atlanıyor.")
        continue

    trend_start = min_race_date - timedelta(days=10)
    trend_end   = max_race_date + timedelta(days=10)

    print(f"\n{y} yılı için trend aralığı: {trend_start.date()} -> {trend_end.date()}")

    # O yılın benzersiz sürücüleri
    riders_y = (
        df_y["Winning rider"]
        .astype(str)
        .str.replace(r"\[.*?\]", "", regex=True)
        .str.strip()
        .unique()
    )

    for rider in sorted(riders_y):
        key = (rider, y)
        if key in rider_trend_data:
            continue  # daha önce çekildiyse tekrar çekme

        print(f"  {y} - Trend serisi çekilen sürücü: {rider}")
        series_df = safe_trend_series(rider, trend_start, trend_end)
        rider_trend_data[key] = series_df
        time.sleep(2)  # istekler arasına ufak buffer

def get_search_interest_from_series(name: str, race_date: datetime, year: int):
    """
    Artık burada pytrends'e istek atmıyoruz.
    Önceden çektiğimiz rider_trend_data[(name, year)] içinden
    7 gün önce / 7 gün sonra ortalamayı alıyoruz.
    """
    if race_date is None:
        return 0, 0, 0, None

    key = (name, year)
    series_df = rider_trend_data.get(key)
    if series_df is None or series_df.empty:
        return 0, 0, 0, None

    s = series_df[name]

    before_start = race_date - timedelta(days=7)
    before_end   = race_date - timedelta(days=1)

    after_start  = race_date + timedelta(days=1)
    after_end    = race_date + timedelta(days=7)

    before_mask = (s.index >= before_start) & (s.index <= before_end)
    after_mask  = (s.index >= after_start) & (s.index <= after_end)

    before_vals = s[before_mask]
    after_vals  = s[after_mask]

    before_mean = before_vals.mean() if not before_vals.empty else 0
    after_mean  = after_vals.mean() if not after_vals.empty else 0

    trend_diff = after_mean - before_mean
    rel_inc = (trend_diff / before_mean * 100) if before_mean > 0 else None

    return before_mean, after_mean, trend_diff, rel_inc

# ---------------------------------------------------------
# 4.1) TÜM YARIŞLAR İÇİN BEFORE/AFTER HESAPLA
# ---------------------------------------------------------

search_before = []
search_after = []
trend_diff_list = []
rel_inc_list = []

print("\nYarış başına BEFORE / AFTER hesaplanıyor (yeni yöntem, pytrends çağrısı yok)...")

for idx, row in motogp_full.iterrows():
    rider = re.sub(r"\[.*?\]", "", str(row["Winning rider"])).strip()
    race_date = row["RaceDate"]
    year = int(row["Year"])

    b, a, d, r = get_search_interest_from_series(rider, race_date, year)

    search_before.append(b)
    search_after.append(a)
    trend_diff_list.append(d)
    rel_inc_list.append(r)

motogp_full["Search Before"]       = search_before
motogp_full["Search After"]        = search_after
motogp_full["Trend Difference"]    = trend_diff_list
motogp_full["Relative Increase %"] = rel_inc_list

# ---------------------------------------------------------
# 5) SON KOLON SIRASI + GÖSTERİM
# ---------------------------------------------------------

col_order = [
    "Year",
    "RaceDate",
    "Date",
    "Grand Prix",
    "Circuit",
    "Winning rider",
    "Win margin",
    "Career wins",
    "Career podiums",
    "Championships",
    "Country",
    "Years active",
    "Career wins_num",
    "Career podiums_num",
    "Championships_num",
    "Years_active_len",
    "Popularity",
    "Search Before",
    "Search After",
    "Trend Difference",
    "Relative Increase %",
]

col_order = [c for c in col_order if c in motogp_full.columns]
motogp_full = motogp_full[col_order]

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
pd.set_option("display.width", 240)
pd.set_option("display.max_colwidth", None)

print("\n--- motogp_full (TÜM SATIRLAR) ---")
print(motogp_full.to_string(index=False))

# İstersen export da ekleyebilirsin:
# motogp_full.to_csv("motogp_full_with_trends_yearbyyear.csv", index=False)
```

### Paired t-test on search interest (before vs. after)

This block performs a **paired t-test** comparing Google search interest *before* and *after* races:

- Filters out rows with missing values
- Runs a **two-sided paired t-test**
- Converts it into a **one-sided p-value** to test whether search interest tends to **increase after races**

The output helps determine if post-race attention is significantly higher than pre-race levels.


```python
from scipy import stats

valid = motogp_full[["Search Before", "Search After"]].dropna()

before = valid["Search Before"]
after  = valid["Search After"]

t_stat, p_two_sided = stats.ttest_rel(after, before)

t_stat, p_two_sided
p_one_sided = p_two_sided / 2 if t_stat > 0 else 1 - p_two_sided / 2
p_one_sided
```

### T-Test: Do Experienced Riders Have Higher Popularity?
This test compares *experienced riders* (>5 active years) with *newer riders* (≤5 years) using Welch’s independent t-test.  
The one-sided hypothesis checks whether long-tenured riders achieve *significantly higher popularity scores*.

The result reports:
- sample sizes for both groups  
- t-statistic and one-sided p-value  
- a conclusion on whether experience meaningfully boosts popularity.

```python
from scipy import stats
import numpy as np

# 1. Data Cleaning: Drop NaN values for the columns of interest
valid_riders = motogp_full.dropna(subset=["Years_active_len", "Popularity"])

# 2. Define Groups: Using the 5-year threshold
EXPERIENCED = valid_riders[valid_riders["Years_active_len"] > 5]["Popularity"].values
NEW = valid_riders[valid_riders["Years_active_len"] <= 5]["Popularity"].values

print(f"Experienced (> 5 years) sample size: {len(EXPERIENCED)}")
print(f"New (<= 5 years) sample size: {len(NEW)}")

# 3. Independent Two-Sample T-Test (using equal_var=False for Welch's T-Test, which is more robust)
# Our alternative hypothesis is one-sided (greater than).
t_stat, p_two_sided = stats.ttest_ind(EXPERIENCED, NEW, equal_var=False)

# 4. Calculate One-Sided P-value
if t_stat > 0:
    p_one_sided = p_two_sided / 2
else:
    # If the Experienced group mean is lower, we reject the one-sided hypothesis.
    p_one_sided = 1 - p_two_sided / 2

print(f"\nIndependent T-Test Results (Experienced vs. New Riders):")
print(f"T-Statistic: {t_stat:.4f}")
print(f"One-Sided P-value: {p_one_sided:.4f}")

# 5. Interpretation
alpha = 0.05
if p_one_sided < alpha:
    print(f"\nConclusion: The P-value ({p_one_sided:.4f}) is below the significance level ({alpha}).")
    print("H0 is rejected. The Popularity score of experienced riders is significantly higher than that of new riders.")
else:
    print(f"\nConclusion: The P-value ({p_one_sided:.4f}) is above the significance level ({alpha}).")
    print("H0 is not rejected. There is no significant difference in Popularity between the groups.")
```

### Champion vs Non-Champion Search Impact (T-Test)

- Data is cleaned and split into two groups: *champions* (≥1 title) and *non-champions* (0 titles).  
- Their *Relative Increase %* values are compared using a Welch two-sample t-test.  
- A one-sided p-value tests whether champions trigger a larger search spike.  
- *If p < 0.05:* champion wins lead to a significantly higher post-race search increase.  
- *If p ≥ 0.05:* no statistically significant difference between the two groups.

```python
from scipy import stats
import numpy as np

# 1. Data Cleaning: Drop NaNs and infinite values for trend and championship columns
valid_trend = motogp_full.dropna(subset=["Championships_num", "Relative Increase %"])
# Filter out 'inf' (infinite) values which can occur if 'Search Before' was 0 (to avoid errors in t-test)
valid_trend = valid_trend[np.isfinite(valid_trend["Relative Increase %"])]

# 2. Define Groups
CHAMPION = valid_trend[valid_trend["Championships_num"] > 0]["Relative Increase %"].values
NON_CHAMPION = valid_trend[valid_trend["Championships_num"] == 0]["Relative Increase %"].values

print(f"\nChampion (> 0 Championships) sample size: {len(CHAMPION)}")
print(f"Non-Champion (= 0 Championships) sample size: {len(NON_CHAMPION)}")

# 3. Independent Two-Sample T-Test (using equal_var=False)
t_stat_c, p_two_sided_c = stats.ttest_ind(CHAMPION, NON_CHAMPION, equal_var=False)

# 4. Calculate One-Sided P-value
if t_stat_c > 0:
    p_one_sided_c = p_two_sided_c / 2
else:
    # If the Champion mean is lower, we reject the one-sided hypothesis.
    p_one_sided_c = 1 - p_two_sided_c / 2

print(f"\nIndependent T-Test Results (Champion vs. Non-Champion):")
print(f"T-Statistic: {t_stat_c:.4f}")
print(f"One-Sided P-value: {p_one_sided_c:.4f}")

# 5. Interpretation
alpha = 0.05
if p_one_sided_c < alpha:
    print(f"\nConclusion: The P-value ({p_one_sided_c:.4f}) is below the significance level ({alpha}).")
    print("H0 is rejected. Victories by champion riders create a significantly larger relative increase in search interest than those by non-champion riders.")
else:
    print(f"\nConclusion: The P-value ({p_one_sided_c:.4f}) is above the significance level ({alpha}).")
    print("H0 is not rejected. There is no significant difference in the relative search interest increase based on championship status.")
```

### Correlation Between Popularity and Relative Search Increase

- Clean data is used to compute a *Pearson correlation* between rider *Popularity* and *Relative Increase %*.  
- The one-sided p-value tests whether higher popularity is associated with a larger post-race search spike.  
- *If p < 0.05:* there is a significant positive correlation — more popular riders tend to generate bigger search increases.  
- *If p ≥ 0.05:* no significant positive relationship is found.

```python
from scipy.stats import pearsonr
import numpy as np

# Data Cleaning: Both columns must have valid data for correlation calculation
valid_data = motogp_full.dropna(subset=["Popularity", "Relative Increase %"])
valid_data = valid_data[np.isfinite(valid_data["Relative Increase %"])]

popularity = valid_data["Popularity"]
relative_increase = valid_data["Relative Increase %"]

# Calculate the Pearson Correlation Coefficient and the 2-sided p-value
# Correlation testing measures the strength and direction of the relationship directly.
correlation_coeff, p_two_sided = pearsonr(popularity, relative_increase)

# One-Sided P-value (We are looking for a positive relationship)
p_one_sided = p_two_sided / 2 if correlation_coeff > 0 else 1 - p_two_sided / 2

print(f"\nPearson Correlation Coefficient: {correlation_coeff:.4f}")
print(f"One-Sided P-value: {p_one_sided:.4f}")

# Interpretation
alpha = 0.05
if p_one_sided < alpha:
    print(f"\nConclusion: The P-value ({p_one_sided:.4f}) is below the significance level ({alpha}).")
    print("H0 is rejected. As the rider's Popularity score increases, the relative search increase after the race also significantly increases.")
else:
    print(f"\nConclusion: The P-value ({p_one_sided:.4f}) is above the significance level ({alpha}).")
    print("H0 is not rejected. There is no significant positive correlation between Popularity and the relative search increase.")
```

### Correlation Between Pre-Race Search Interest and Relative Increase

- Computes Pearson correlation between *Search Before* and *Relative Increase %*, using clean and finite data only.  
- One-sided test checks whether higher baseline interest leads to a *smaller* post-race spike (negative correlation).  
- *If p < 0.05:* baseline search interest significantly predicts a smaller relative increase after the race.  
- *If p ≥ 0.05:* no significant negative relationship is detected.

```python
from scipy.stats import pearsonr
import numpy as np
import pandas as pd

# --- 1. Data Cleaning ---
# Filter out rows with missing values in the variables of interest
valid_corr_data_2 = motogp_full.dropna(subset=["Search Before", "Relative Increase %"])
valid_corr_data_2 = valid_corr_data_2[np.isfinite(valid_corr_data_2["Relative Increase %"])]

search_before = valid_corr_data_2["Search Before"]
relative_increase_2 = valid_corr_data_2["Relative Increase %"]

# --- 2. Correlation Calculation ---
correlation_coeff_2, p_two_sided_2 = pearsonr(search_before, relative_increase_2)

# One-Sided P-value (Testing for a negative relationship: r < 0)
p_one_sided_2 = p_two_sided_2 / 2 if correlation_coeff_2 < 0 else 1 - p_two_sided_2 / 2

print(f"\nPearson Correlation Coefficient (Search Before vs Relative Increase %): {correlation_coeff_2:.4f}")
print(f"One-Sided P-value: {p_one_sided_2:.4f}")

# --- 3. Interpretation ---
alpha = 0.05
if p_one_sided_2 < alpha:
    print(f"\nResult: The P-value ({p_one_sided_2:.4f}) is below the significance level ({alpha}).")
    print("H0 is rejected. As the rider's pre-race search interest (Search Before) increases, the relative post-race search increase (Relative Increase %) significantly decreases.")
else:
    print(f"\nResult: The P-value ({p_one_sided_2:.4f}) is above the significance level ({alpha}).")
    print("H0 is not rejected. There is no significant negative correlation between the baseline interest level and the relative search spike.")
```
